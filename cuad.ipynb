```python
import os

### Veremos todas las carpetas y subcarpetas del directorio /
## Diseñamos un print intersante del tipo:
## proyecto/
#├── app/
#│   ├── requirements.txt
#│   └── app.py

def listar_directorio(directorio, nivel=0):
    for elemento in os.listdir(directorio):
        ruta = os.path.join(directorio, elemento)
        print("│   " * nivel + "├── " + elemento+"\n")
        if os.path.isdir(ruta):
            listar_directorio(ruta, nivel + 1)

directorio = "/home/isaac-zainea/Dropbox/Documentos/EAN/Nuevo Proyecto"
print(directorio)
listar_directorio(directorio)

# Si queremos ver los archivos de un directorio en particular
directorio = "."
print(directorio)
```
